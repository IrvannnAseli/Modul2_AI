**Bagian 1: Object-Oriented Programming (OOP)**

**1.1 Definisi Class dan Object**

```python
class Mobil:
    def __init__(self, merk, model, tahun):
        self.merk = merk
        self.model = model
        self.tahun = tahun

    def tampilkan_info(self):
        print(f"Mobil: {self.merk} {self.model}, Tahun: {self.tahun}")

# Contoh penggunaan
mobil1 = Mobil("Toyota", "Avanza", 2020)
mobil1.tampilkan_info()
```

**1.2 Inheritance (Pewarisan)**

```python
class Manusia:
    def __init__(self, nama, umur):
        self.nama = nama
        self.umur = umur

class Pelajar(Manusia):
    def __init__(self, nama, umur, sekolah):
        super().__init__(nama, umur)
        self.sekolah = sekolah

    def tampilkan_sekolah(self):
        print(f"{self.nama} bersekolah di {self.sekolah}")

# Contoh penggunaan
pelajar1 = Pelajar("Ali", 16, "SMK Negeri 1")
pelajar1.tampilkan_sekolah()
```

**Bagian 2: Error Handling dan Exception**

**2.1 Penanganan Error dengan Try-Except**

```python
try:
    angka = int(input("Masukkan sebuah angka: "))
    hasil = 10 / angka
    print(f"Hasil: {hasil}")
except ZeroDivisionError:
    print("Tidak bisa membagi dengan nol!")
except ValueError:
    print("Input harus berupa angka!")
finally:
    print("Program selesai.")
```

**Bagian 4: Lambda Function dan List Comprehension**

**4.1 Lambda Function**

```python
# Fungsi lambda untuk penjumlahan dua angka
penjumlahan = lambda x, y: x + y
print(penjumlahan(5, 3))
# Menggunakan lambda dalam filter
angka = [1, 2, 3, 4, 5, 6]
ganjil = list(filter(lambda x: x % 2 != 0, angka))
print(ganjil)
```

**4.2 List Comprehension**

```python
# Membuat list baru dari list yang ada
angka = [1, 2, 3, 4, 5]
kuadrat = [x**2 for x in angka]
print(kuadrat)
# Menggunakan kondisi
ganjil = [x for x in angka if x % 2 != 0]
print(ganjil)
```

**Bagian 5: Regular Expression (Regex)**

**5.1 Pengenalan Regex**

```python
import re

# Mencari pola
teks = "Email saya adalah riski@example.com"
pola = r'\w+@\w+\.\w+'
hasil = re.search(pola, teks)

if hasil:
    print("Email ditemukan:", hasil.group())
```

```python
# Class Kalkulator
class Kalkulator:
    def tambah(self, a, b):
        return a + b

    def kurang(self, a, b):
        return a - b

    def kali(self, a, b):
        return a * b

    def bagi(self, a, b):
        if b == 0:
            return "Tidak bisa membagi dengan nol"
        return a / b

# Contoh penggunaan
kalkulator = Kalkulator()
print("Tambah:", kalkulator.tambah(10, 5))
print("Kurang:", kalkulator.kurang(10, 5))
print("Kali:", kalkulator.kali(10, 5))
print("Bagi:", kalkulator.bagi(10, 5))
```

**5. Regular Expression**

```python
import re

teks = "Kontak kami di support@example.com atau admin@example.org untuk informasi lebih lanjut."
email_regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
email_list = re.findall(email_regex, teks)
print("Email yang ditemukan:", email_list)
```

```python
nomor_telepon = "0812-3456-7890"
telepon_regex = r'^08\d{2}-\d{4}-\d{4}$'

if re.match(telepon_regex, nomor_telepon):
    print("Nomor telepon valid.")
else:
    print("Nomor telepon tidak valid.")
```
